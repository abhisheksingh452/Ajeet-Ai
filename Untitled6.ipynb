```python
import random
from collections import deque
import matplotlib.pyplot as plt

# BFS shortest path distance
def bfs(a, b):
    q = deque([(a, 0)])
    v = {a}
    moves = [(1,0),(-1,0),(0,1),(0,-1)]
    while q:
        (x,y), d = q.popleft()
        if (x,y) == b: return d
        for dx,dy in moves:
            nx, ny = x+dx, y+dy
            if 0<=nx<10 and 0<=ny<10 and (nx,ny) not in v:
                v.add((nx,ny))
                q.append(((nx,ny), d+1))
    return 999

# Random garbage bins
bins = [(random.randint(0,9), random.randint(0,9)) for _ in range(6)]

# Divide bins between 2 agents
A_bins = bins[:3]
B_bins = bins[3:]

A_pos = (0,0)
B_pos = (0,0)
A_dist = 0
B_dist = 0

# Travel & count distance
for b in A_bins:
    A_dist += bfs(A_pos, b)
    A_pos = b

for b in B_bins:
    B_dist += bfs(B_pos, b)
    B_pos = b

print("Garbage Bins:", bins)
print("\nAgent A bins:", A_bins)
print("Agent B bins:", B_bins)
print("\nDistance A:", A_dist)
print("Distance B:", B_dist)

# Simple plot
for x,y in bins:
    plt.scatter(x,y,c='red')

plt.scatter(0,0,c='blue',label='Depot')
plt.title("Garbage Collection Routing")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Traffic Light Agent
class TrafficLightAgent:
    def __init__(self, name, min_green=5, max_green=30):
        self.name = name
        self.min_green = min_green
        self.max_green = max_green
        self.green_time = 10
        self.wait_log = []

    def decide_green_time(self, queue_length):
        if queue_length > 20:
            self.green_time += 5
        elif queue_length < 5:
            self.green_time -= 3

        self.green_time = max(self.min_green, min(self.green_time, self.max_green))
        return self.green_time

    def record_wait(self, queue):
        self.wait_log.append(queue)

# Environment
class TrafficEnvironment:
    def __init__(self):
        self.NS_queue = random.randint(10, 30)
        self.EW_queue = random.randint(10, 30)

        self.NS_agent = TrafficLightAgent("NS-Agent")
        self.EW_agent = TrafficLightAgent("EW-Agent")

        self.logs = {"NS": [], "EW": []}

    def simulate_cycle(self):
        ns_green = self.NS_agent.decide_green_time(self.NS_queue)
        ew_green = self.EW_agent.decide_green_time(self.EW_queue)

        self.NS_queue = max(0, self.NS_queue - random.randint(5, 12))
        self.EW_queue = max(0, self.EW_queue - random.randint(5, 12))

        self.NS_agent.record_wait(self.NS_queue)
        self.EW_agent.record_wait(self.EW_queue)

        self.logs["NS"].append(self.NS_queue)
        self.logs["EW"].append(self.EW_queue)

    def run_simulation(self, cycles=20):
        for _ in range(cycles):
            self.simulate_cycle()
        return self.logs

# Plot result
def plot_results(logs):
    plt.plot(logs["NS"], label="NS Queue", linewidth=3)
    plt.plot(logs["EW"], label="EW Queue", linewidth=3)
    plt.xlabel("Cycle")
    plt.ylabel("Queue Length")
    plt.title("Traffic Agent Performance")
    plt.legend()
    plt.grid(True)
    plt.show()

# Run simulation
env = TrafficEnvironment()
logs = env.run_simulation()
plot_results(logs)

print("Final Average NS Waiting:", sum(logs["NS"]) / len(logs["NS"]))
print("Final Average EW Waiting:", sum(logs["EW"]) / len(logs["EW"]))
```
