```python
# Grid Painting Agents – Simple Version
# Two agents paint a grid without overlapping using DFS-like logic.

import numpy as np
import matplotlib.pyplot as plt

rows, cols = 8, 8
grid = np.zeros((rows, cols), dtype=int)

# Agent positions
a1 = [0, 0]              # Agent 1 start
a2 = [rows-1, cols-1]    # Agent 2 start

grid[a1[0]][a1[1]] = 1
grid[a2[0]][a2[1]] = 2

# DFS stacks
stack1 = [tuple(a1)]
stack2 = [tuple(a2)]

# Neighbour directions
moves = [(-1,0),(1,0),(0,-1),(0,1)]

def dfs_step(pos, stack, agent_id):
    r, c = pos
    moved = False

    # Try to move to any unpainted neighbour
    for dr, dc in moves:
        nr, nc = r+dr, c+dc
        if 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] == 0:
            stack.append((r, c))
            grid[nr][nc] = agent_id
            return [nr, nc], stack

    # No move possible → backtrack
    if stack:
        return list(stack.pop()), stack
    else:
        return pos, stack


# Run both agents
for _ in range(300):
    a1, stack1 = dfs_step(a1, stack1, 1)
    a2, stack2 = dfs_step(a2, stack2, 2)

# Plot result
plt.imshow(grid, cmap='tab20')
plt.title("Grid Painting by Two Agents")
plt.colorbar()
plt.show()
```

```python
# Resource Collection Team – Simple Version
# Multiple agents collect resources using a shared task queue.

import random
import numpy as np
import matplotlib.pyplot as plt
from collections import deque

rows, cols = 10, 10
num_resources = 20
num_agents = 3

# Create grid with resources
grid = np.zeros((rows, cols), dtype=int)

count = 0
while count < num_resources:
    r, c = random.randint(0,rows-1), random.randint(0,cols-1)
    if grid[r][c] == 0:
        grid[r][c] = 1
        count += 1

# Shared task queue
tasks = deque([(r, c) for r in range(rows) for c in range(cols) if grid[r][c] == 1])

# Agent positions & collected counts
agents = [[random.randint(0,rows-1), random.randint(0,cols-1)] for _ in range(num_agents)]
collected = [0]*num_agents
targets = [None]*num_agents


def move_towards(pos, target):
    r, c = pos
    tr, tc = target

    if tr > r: return [r+1, c]
    if tr < r: return [r-1, c]
    if tc > c: return [r, c+1]
    if tc < c: return [r, c-1]
    return pos  # already at target


for _ in range(300):
    for i in range(num_agents):

        # If no target, take one from queue
        if targets[i] is None and tasks:
            targets[i] = tasks.popleft()

        # If still no target → idle
        if targets[i] is None:
            continue

        # Move toward target
        agents[i] = move_towards(agents[i], targets[i])

        # If reached the target
        if agents[i] == list(targets[i]):
            r, c = targets[i]
            if grid[r][c] == 1:
                grid[r][c] = 0
                collected[i] += 1
            targets[i] = None


# Plot performance
plt.bar([f"Agent {i+1}" for i in range(num_agents)], collected)
plt.title("Resources Collected by Each Agent")
plt.ylabel("Count")
plt.show()
```
